```python
import numpy as np
import cv2 as cv
import skimage.io as io
import os
import matplotlib.pyplot as plt
```

```python

ls= os.listdir("preds") 

for i in ls:
    if i[-1]=="g":
        test= io.imread("preds/"+i)
        plt.imshow(test, cmap="gray")
        plt.show()
        
        kernel= cv.getStructuringElement(cv.MORPH_RECT, (20,20))
        closing= cv.morphologyEx(test, cv.MORPH_OPEN, kernel)
        kernel2= cv.getStructuringElement(cv.MORPH_RECT, (20,20))
        opening= cv.morphologyEx(closing, cv.MORPH_CLOSE, kernel2)

        #closing=closing/closing.max()
        #closing=closing/255
        #closing[closing>0.5]=1
        #closing[closing<0.5]=0


        io.imsave(os.path.join("preds_closed",i), closing)
        plt.imshow(opening, cmap="gray")
        plt.show()
```
